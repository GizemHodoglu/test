```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data.csv')
df = df.rename(columns={'Item1': 'Meteorolojiye Dair Bilgim', 'Item2': 'Havanın Tahmin Edilebilirliğine İnancım'})
ax = df.plot(x='Yil', y=['Meteorolojiye Dair Bilgim', 'Havanın Tahmin Edilebilirliğine İnancım'])
ax.set_ylabel("Bilgi")
ax.set_xlabel("Yıl")
ax.tick_params(axis='both', which='both', length=0, labelbottom=False, labelleft=False)
plt.show()
```
